```python
# Carregar o dataset usando Hugging Face Datasets
dataset = load_dataset("text", data_files={"train": "frases_pro_meu_amor.txt"})

tokenizer.add_special_tokens({'pad_token': '[PAD]'})
tokenizer.pad_token = tokenizer.eos_token

# Preparar os dados para o treinamento
def tokenize_data(examples):
    return tokenizer(examples['text'], return_tensors='pt', truncation=True, padding=True, max_length=128)

tokenized_dataset = dataset.map(tokenize_data, batched=True, remove_columns=['text'])

data_collator = DataCollatorForLanguageModeling(
      tokenizer=tokenizer,
      mlm=False
  )

# Configurar os parâmetros de treinamento
training_args = TrainingArguments(
    output_dir="./romantic_gpt2",
    overwrite_output_dir=True,
    num_train_epochs=5,
    per_device_train_batch_size=6,
    save_steps=500,
    save_total_limit=2,
    run_name="Novos_encontros(15/02/25)/1"
)

# Iniciar o treinamento
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_dataset['train']
)

trainer.train()
```

```python
# Salvar o modelo treinado
trainer.save_model("./romantic_gpt2")
```

```python
# Função para gerar resposta
def gerar_resposta(input_text):
    # Codifica o texto de entrada
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # Criar attention mask
    attention_mask = input_ids != tokenizer.pad_token_id

    # Gera o texto de resposta
    output = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=50,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.8,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decodifica o texto gerado
    resposta = tokenizer.decode(output[0], skip_special_tokens=True)
    return resposta

# Testando a resposta
# Loop de interação com o usuário
while True:
    entrada_usuario = input("Você: ")  # Solicita entrada do usuário
    if entrada_usuario.lower() == "sair":  # Permite sair do loop
        break
    resposta_modelo = gerar_resposta(entrada_usuario)  # Gera resposta do modelo
    print("Chatbot:", resposta_modelo)  # Exibe a resposta do modelo

print("Chatbot: Até logo!")
```
